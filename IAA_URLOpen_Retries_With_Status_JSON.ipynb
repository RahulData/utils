```python
def url_open_retries(url):
    tries = 4
    delay = 3
    backoff = 1
    mtries, mdelay = tries, delay
    image_bytes = ''
    
    while mtries > 0:
        try:
            print(mtries)
            response = request.urlopen(url)
            image_bytes = response.read()
            print('Success. Exiting..')
            fn_status = True
            error_type = 'None'
            break
        except (ValueError, request.HTTPError, request.URLError) as ex:
            print('Exception Type: ' + type(ex).__name__ +' .. Retrying..')
    #         msg = "%s, Retrying in %d seconds..." % (str(ex), mdelay)
            time.sleep(mdelay)
            mtries -= 1
            mdelay *= backoff
            fn_status = False
            error_type = type(ex).__name__
    
    print('END')
    return fn_status, image_bytes, error_type
```

```python
input_stocks_list = df1[col_list].sample(n=1)

status_json = {}

status_json['123'] = {'Status':'SUCCESS','Error':'NONE'}
    
for index, el in input_stocks_list.iterrows(): 
#     print (el["stockNumber"], el["url01"]) 
    el_stock = el['stockNumber']
    for img_no in range(1,11):
        img_column = str(img_no)  
        dyn_col = 'url0'+img_column
        if(img_no == 10):
            dyn_col = 'url'+img_column
        el_url = el[dyn_col]  # e..g. 'url10'
        el_img_no = dyn_col[3:5]

        el_url_open_status, el_img_bytes, el_error_type = url_open_retries(el_url)
        print('Stock:' + str(el_stock) + ' Status: '+ str(el_url_open_status))
        if(el_url_open_status):
            el_status = 'SUCCESS...'
        status_json[el_stock] = {'Status':el_status,'URL':dyn_col,'Error':el_error_type}
#         if(el_url_open_status):
#             el_status = download_img_bytes_to_blob(block_blob_service, container_name, el_stock, el_img_no, el_img_bytes)
#             if (el_status):
#                 processed_count += 1
    
# for el in input_stocks_list:
#     el_stock = el['stockNumber']
#     for img_no in range(1,11):
#         img_column = str(img_no)  
#         dyn_col = 'url0'+img_column
#         if(img_no == 10):
#             dyn_col = 'url'+img_column
#         el_url = el[dyn_col]  # e..g. 'url10'
#         el_img_no = dyn_col[3:5]

#         el_url_open_status, el_img_bytes, el_error_type = url_open_retries(el_url)
#         print('Stock:' + str(el_stock) + ' Status: '+el_url_open_status)
# #         if(el_url_open_status):
# #             el_status = download_img_bytes_to_blob(block_blob_service, container_name, el_stock, el_img_no, el_img_bytes)
# #             if (el_status):
# #                 processed_count += 1
```
