### Setting labels, legends, and other custom properties in Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({'X':[1,2,3,4,5,6,7,8],'Y':[3,5,7,8,13,7,5,9],'Z':[2,4,6,7,14,6,4,8]})
df
```

```python
plt.figure(figsize=(10,6))
plt.rc('axes', titlesize=16)  # fontsize of the axes title
plt.rc('axes', labelsize=16) # fontsize of the x and y labels

plt.plot(df['X'], df[['Y','Z']], lw=3) # lw = line width/thickness
plt.legend(['Actual','Pred'])
plt.xlabel('Weeks')
plt.ylabel('Sales')
plt.title('Sales by Week (''000 $)')
plt.show()

```
